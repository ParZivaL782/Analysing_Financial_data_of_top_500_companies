```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df=pd.read_csv("Financial Analytics data.csv")
```

```python
df
```

```python
# @title Sales Qtr - Crore

from matplotlib import pyplot as plt
```

```python
df.describe()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.drop(df.columns[4],axis=1,inplace=True)
```

```python
plt.scatter(df["Mar Cap - Crore"], df["Sales Qtr - Crore"])
plt.xlabel("Market Capitalization (crores)")
plt.ylabel("Quarterly Sales (crores)")
plt.title("Scatter Plot of Market Cap vs Quarterly Sales")
plt.show()
```

```python
df['Mar Cap - Crore'].plot(kind='line', figsize=(8,4), title='Mar Cap - Crore')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df['Sales Qtr - Crore'].plot(kind='line', figsize=(8, 4), title='Sales Qtr - Crore')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
correlation = df["Mar Cap - Crore"].corr(df["Sales Qtr - Crore"])
print(f"Correlation between Market Cap and Quarterly Sales: {correlation:.2f}")
```

```python
correlation = np.reshape(correlation, (1, 1))
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.show()
```
